# Importing Librabries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Heart1.csv")
```

```python
#a) read data from different formats (like csv, xls)
df.head()
```

# Bar Plot:

```python
# Bar Plot - Chest Pain Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='ChestPain', hue='AHD', data=df, palette='Set1')
plt.title('Bar Plot: Count of Chest Pain Types')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.xticks(rotation=30)
plt.show()
```

# Scatter Plot:

```python
# Scatter Plot - Age vs Cholesterol
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Age"], y=df["Chol"], hue=df["AHD"], alpha=0.7, palette="Set1")
plt.title("Scatter Plot: Age vs Cholesterol")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.show()
```

# Box Plot:

```python
# Box Plot - Resting Blood Pressure by AHD
plt.figure(figsize=(8, 6))
sns.boxplot(x=df["AHD"], y=df["RestBP"], palette="muted")
plt.title("Box Plot: Resting Blood Pressure by AHD", fontsize=14)
plt.xlabel("AHD (Heart Disease)", fontsize=12)
plt.ylabel("Resting Blood Pressure", fontsize=12)
plt.show()
```

# Pie Chart:

```python
# Pie Chart - Sex Distribution
sex_counts = df["Sex"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(sex_counts, labels=["Male", "Female"], autopct='%1.1f%%', colors=["#ff9999", "#66b3ff"], startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title("Pie Chart: Distribution of Sex", fontsize=14)
plt.show()
```

# Line Chart:

```python
# Line Chart - Max Heart Rate vs Age
plt.figure(figsize=(8, 6))
sns.lineplot(x=df["Age"], y=df["MaxHR"], hue=df["AHD"], palette="tab10", marker="o")
plt.title("Line Chart: Max Heart Rate vs Age", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Maximum Heart Rate", fontsize=12)
plt.show()
```

# Histogram:

```python
# Histogram - Age Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df["Age"], bins=10, kde=True, color="mediumseagreen")
plt.title("Histogram: Age Distribution", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

# Heatmap:

```python
# Heatmap - Correlation Matrix
plt.figure(figsize=(10, 8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap: Correlation Matrix", fontsize=14)
plt.show()
```

# Violin Plot:

```python
# Violin Plot - Cholesterol Distribution by AHD
plt.figure(figsize=(8, 6))
sns.violinplot(x=df["AHD"], y=df["Chol"], palette="Set3")
plt.title("Violin Plot: Cholesterol Distribution by AHD", fontsize=14)
plt.xlabel("AHD (Heart Disease)", fontsize=12)
plt.ylabel("Cholesterol", fontsize=12)
plt.show()
```

# Bubble Plot:

```python
# Bubble Plot - Age vs Cholesterol (Oldpeak as Bubble Size)
plt.figure(figsize=(10, 8))
plt.scatter(df["Age"], df["Chol"], s=df["Oldpeak"]*100, alpha=0.6, c="orange", edgecolors="black")
plt.title("Bubble Plot: Age vs Cholesterol (Bubble Size = Oldpeak)", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Cholesterol", fontsize=12)
plt.show()
```
